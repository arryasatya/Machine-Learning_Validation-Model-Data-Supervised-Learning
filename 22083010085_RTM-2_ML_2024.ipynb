```python
import pandas as pd
import numpy as np
from scipy.stats import mode
```

```python
iris = pd.read_csv('Data_Iris.csv', header=None)
iris.columns = ['x1', 'x2', 'x3', 'x4', 'class']
iris.index = range(1, len(iris) + 1)
```

```python
iris
```

```python
# Kelompokkan data berdasarkan kelasnya
grouped_data = iris.groupby('class')

# Inisialisasi list untuk menyimpan data pelatihan dan data testing
train_data = []
test_data = []

# Ambil 10 data terakhir dari setiap kelas sebagai data testing
for _, group in grouped_data:
    test_data.append(group.tail(10))
    train_data.append(group.iloc[:-10])

# Gabungkan data pelatihan dan data testing
train_data = pd.concat(train_data)
test_data = pd.concat(test_data)

class_mapping = {'Iris-setosa': 1, 'Iris-versicolor': 2, 'Iris-virginica': 3}

train_data['label'] = train_data['class'].map(class_mapping)
test_data['label'] = test_data['class'].map(class_mapping)

train_data.drop(columns=['class'], inplace=True)
test_data.drop(columns=['class'], inplace=True)
```
